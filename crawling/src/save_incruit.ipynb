```python
import os
import docx2txt
import pandas as pd

#custom
from docPreprocessor import DocxDocument
```

```python
path = "../../datafiles/raw-data"

os.listdir(path)
```

```python
for data_dir in os.listdir(path):
    if data_dir[0] == '.':
        continue
    for file in os.listdir(path + "/" + data_dir):
        ext = file[-3:]
        # ms word file
        if ext == 'ocx':
            text = docx2txt.process(path+"/"+file)
            lines = []
            for line in text.splitlines():
                if line == '\n' or line == '':
                    continue
                lines.append(line)
                
            docxObj = DocxDocument(lines)
            corpName = docxObj.extractCorpName()
            task = docxObj.extractTask()
            q, a = docxObj.extractQAList()

            task_list.append(task)
            corp_list.append(corpName)
            question_list.append(q)
        # hwp file
        elif ext == 'hwp':
            filename = "hwp/"+file[:-4]+".txt"
            output = '--output ' + '"' + filename + '"'
            exefile = 'hwp5txt'
            result = '"' + path+'/'+file + '"'
            os.system(exefile + " " + output + " " + result)
        # pdf file
        elif ext == 'pdf':
            pdObj = PdfDocument(path + "/" + file)
            q, a = pdObj.extractQAList()

            for i in range(len(q)):
                tmp_qst.append(q[i])
                tmp_ans.append(a[i])
        question_list.append(tmp_qst)
```
